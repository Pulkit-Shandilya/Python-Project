# Predicting price of Houses based on area of house, implementation of Simple linear Regression through SciLearn.

## for the following dataset,

|     Area (m2) | 100 | 300 | 500 | 800 | 1000 | 1500 |
|---------------|-----|-----|-----|-----|------|------|
| Price (lakhs) |  3  |  8  | 10  | 15  |  20  |  40  |



 - What will be regression line equation?

 - price = intercept+slope*area

 - Predict the price of house when area = 1200, using the regression line equation. 


```python
# %pip install scikit-learn

import numpy as np
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt

area = np.array([100,300,500,800,1000,1500]).reshape(-1, 1) 
price = np.array([3,8,10,15,20,40])                    

# Create and train the model
model = LinearRegression()
model.fit(area, price)

Predicted_Area_input= 1200

# Predict price for a new area
new_area = np.array([[Predicted_Area_input]])
predicted_price = model.predict(new_area)

print(f'Predicted price for {Predicted_Area_input} mÂ²: {predicted_price[0]:.2f} lakhs')


# Plotting
plt.scatter(area, price, color='blue', label='Data')
plt.plot(area, model.predict(area), color='red', label='Regression Line')
plt.scatter(new_area, predicted_price, color='green', label='Prediction')
plt.xlabel('Area (mÂ²)')
plt.ylabel('Price (in lakhs)')
plt.title('Simple Linear Regression: Area (mÂ²)vs Price of house')
plt.legend()
plt.grid(True)
plt.show()
```
